$ Y $ mit $ \mu_y = 85, \sigma_y = 18 $

$ \underline{\underline{E(X+2Y)}} = E(X) + E(2Y) = E(X) + 2 \cdot E(Y) = \mu_x + 2\mu_y = 40 + 2 \cdot 85 = \underline{\underline{210}} $

$ \underline{\underline{Var(X+2Y)}} = Var/X) + 2^2 \cdot Var(Y) = \sigma_x^2 + 4 \cdot \sigma_y^2 = 15^2 + 4 \cdot 18^2 = \underline{\underline{1521}} $

$ \underline{\underline{E(X^2)}} = Var(X) + E(X)^2 = \sigma_x^2 + \mu_x^2 = 15^2 + 40^2 = \underline{\underline{1825}} $

$ \underline{\underline{E(U)}} = E(2X+2Y) = E(2X) + E(2Y) = 2 \cdot E(X) + 2 \cdot E(Y) = 2\mu_x + 2\mu_y = 2 \cdot 1000mm + 2 \cdot 500mm = \underline{\underline{3000mm}} $

$ \sigma_U = \sqrt{Var(U)} $

$ Var(U) = Var(2X) + Var(2Y) $

$ \underline{\underline{\sigma_U}} = \sqrt{Var(2X) + Var(2Y)} = \sqrt{2^2 \cdot Var(X) + 2^2 \cdot Var(Y)} = \sqrt{4 \cdot \sigma_x^2 + 4 \cdot \sigma_y^2} = \sqrt{4 \cdot (\sigma_x^2 + \sigma_y^2} = \sqrt{4 \cdot ((0.02mm)^2 + (0.01mm)^2)} = \underline{\underline{0.0447mm}} $

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import norm

matplotlib.rcParams['figure.figsize'] = [16, 8]

n = 5   # Grösse einer Stichprobe
m = 500 # Stichproben

ran = np.array(norm.rvs(size=n*m))
sim = ran.reshape((n, m))
```

## a)

```python
plt.plot(sim)
plt.show()
```

## b)

```python
plt.hist(sim.T, bins=20, density=True, edgecolor='black', facecolor='white')
x = np.linspace(-4, 4, num=100)
y = norm.pdf(x)
plt.plot(x, y)
plt.title('Histogramm sim')
plt.show()

sim_mean = sim.mean(axis=0)
plt.hist(sim_mean, density=True, edgecolor='black', facecolor='white')
y = norm.pdf(x, loc=0, scale=1/np.sqrt(n))
plt.plot(x, y)
plt.title('Histogramm sim_mean')
plt.show()
```

```python
m = 500
for n in [2, 10, 100]:
    sim = np.array(norm.rvs(size=n*m)).reshape((n, m))
    plt.hist(sim.T, bins=20, density=True, edgecolor='black', facecolor='white')
    x = np.linspace(-4, 4, num=100)
    y = norm.pdf(x)
    plt.plot(x, y)
    plt.title('Histogramm sim (n={})'.format(n))
    plt.show()

    sim_mean = sim.mean(axis=0)
    plt.hist(sim_mean, density=True, edgecolor='black', facecolor='white')
    y = norm.pdf(x, loc=0, scale=1/np.sqrt(n))
    plt.plot(x, y)
    plt.title('Histogramm sim_mean (n={})'.format(n))
    plt.show()
    
```

## a)

$ \mu = 100 \Omega, \sigma = 10 \Omega, R \sim \mathcal{N}(\mu, \sigma^2), n=25, x=95 \Omega, P(\overline{x})=? $

$ \sigma_{\overline{X_n}} = \frac{\sigma_X}{\sqrt{n}} = \frac{10}{\sqrt{25}} = 2$

```python
from scipy.stats import norm
from math import sqrt

mean = 100
std = 10
n = 25
x = 95

stderr_x = std / sqrt(n)
p = norm.cdf(x, loc=mean, scale=stderr_x)
print('P(x<95)={:.3f}'.format(p))
```

## b)

$ X \sim \mathcal{U}[4, 6], n = 40, {\overline{X}_{40}}=? $

$ \mu_X = \frac{a+b}{2} = \frac{4 + 6}{2}=5 $

$ \sigma^2_X = Var(X) = \frac{(b-a)^2}{12} = \frac{(6-4)^2}{12} = \frac{1}{3} $

$ \overline{X}_{40} \sim \mathcal{N}(\mu_{\overline{X}_{40}}, \sigma_{\overline{X}_{40}}^2) $

$ \mu_{\overline{X}_{40}} = \mu_X = 5 $

$ \sigma_{\overline{X}_{40}} = \frac{\sigma_X}{\sqrt{n}} = \frac{\sqrt{\frac{1}{3}}}{\sqrt{40}} = \frac{1}{\sqrt{3 \cdot 40}} = \frac{1}{\sqrt{120}} $

$ \overline{X}_{40} \sim \mathcal{N}(5, (\frac{1}{\sqrt{120}})^2) = \mathcal{N}(5, \frac{1}{120}) $

$ \{X_i\}_{1 \leq i \leq 50} \sim \mathcal{N}(\mu, \sigma^2) $ (i.i.d.)

$ \mu = 1, \sigma = 2 $

$ S_n = X_1 + X_2 + \cdots + X_n $

$ \overline{X}_n = \frac{1}{n}(X_1 + X_2 + \cdots + X_n)=\frac{S_n}{n}, n = 50 $

## a)

$ \overline{X}_n \sim \mathcal{N}(\mu_{\overline{X}_n}, \sigma_{\overline{X}_n}^2) $

$ \sigma_{\overline{X}_n} = \frac{\sigma_X}{\sqrt{n}}, \sigma_{\overline{X}_n}^2 = \frac{\sigma_X^2}{n} = \frac{2^2}{50} = 0.08 $

$ \mu_{\overline{X}_n} \mu_X = 1 $

$ \underline{\underline{\overline{X}_n \sim \mathcal{N}(1, 0.08)}} $

$ \underline{\underline{S_n}} \sim \mathcal{N}(n\mu, n\sigma^2_x) = \mathcal{N}(50 \cdot 1, 50 \cdot 2^2) = \underline{\underline{\mathcal{N}(50, 200)}} $

## b)

$ X_1 \sim \mathcal{N}(1, 4) $

$ P(0 \leq X_1 \leq 2) $

```python
from scipy.stats import norm

mean = 1
std = 2
p = norm.cdf(mean+1, loc=mean, scale=std) - norm.cdf(mean-1, loc=mean, scale=std)
print('P(0<=X1<=2)={:.3f}'.format(p))
```

## c)

Zentraler Grenzwertsatz: $ S_n \sim \mathcal{N}(n\mu, n\sigma_x^2) $

$ E[S_n] = n\mu $

```python
from math import sqrt

n = 50
mean = 50
std = sqrt(200)

p = norm.cdf(mean+1, loc=mean, scale=std) - norm.cdf(mean-1, loc=mean, scale=std)
print('P(0<=Sn<=2)={:.3f}'.format(p))
```

# 4.5

$ M_A = \{79.98, 80.04, 80.02, 80.04, 80.03, 80.03, 80.04, 79.97, 80.05, 80.03, 80.02, 80.00, 80.02\} $

$ M_B = \{80.02, 79.94, 79.98, 79.97, 80.03, 79.95, 79.97\} $

Freigesetzte Wärme in $ cal/g $ beim Übergang von Eis ($ -0.7° C $) zu Wasser ($ 0° C $)

## a)

Standardfehler = absoluter Fehler:

$$ \overline{x}_n \pm s_{\overline{x}_n} = \overline{x}_n \pm \frac{s_x}{\sqrt{n}} $$

relativer Fehler:

$$ \overline{x}_n \pm \frac{s_{\overline{x}_n}}{\overline{x}_n} $$

$ s_x $: empirische Standardabweichung der Messreihe:

$$ s_x = \sqrt{\frac{1}{n-1} \sum_{i=1}^n(x_i - \overline{x})^2} $$

```python
import numpy as np
from math import sqrt

ma = np.array([79.98, 80.04, 80.02, 80.04, 80.03, 80.03, 80.04, 79.97, 80.05, 80.03, 80.02, 80.00, 80.02])
mb = np.array([80.02, 79.94, 79.98, 79.97, 79.97, 80.03, 79.95, 79.97])

mean_a = ma.mean()
mean_b = mb.mean()
print('mean a = {:.3f}, mean b = {:.3f}'.format(mean_a, mean_b))

sx_a = sqrt((1 / (len(ma)-1)) * np.sum((ma - mean_a) ** 2))
sx_b = sqrt((1 / (len(mb)-1)) * np.sum((mb - mean_b) ** 2))
print('sxa={:.3f}, sxb={:.3f}'.format(sx_a, sx_b))

abs_err_a = sx_a / sqrt(len(ma))
abs_err_b = sx_b / sqrt(len(mb))
print('manually: abs_err_a={:.5f}, abs_err_b={:.5f}'.format(abs_err_a, abs_err_b))

# using NumPy: ma.std() / np.sqrt(len(ma))
abs_err_a = ma.std(ddof=1) / sqrt(len(ma))
abs_err_b = mb.std(ddof=1) / sqrt(len(mb))
print('NumPy: abs_err_a={:.5f}, abs_err_b={:.5f}'.format(abs_err_a, abs_err_b))
```

Messung A: $ (80.02 \pm 0.01)cal/g $

Messung B: $ (79.98 \pm 0.01)cal/g $

(Messwerte und absoluter Fehler mit gleicher Anzahl Nachkommastellen!)

## b)

absoluter Fehler:

$$ s_{\overline{x}_n} = \frac{s_x}{\sqrt{n}} $$

relativer Fehler:

$$ = \frac{s_{\overline{x}_n}}{\overline{x}_n} $$

```python
rel_err_a = abs_err_a / ma.mean()
rel_err_b = abs_err_b / mb.mean()

print('rel_err_a={:.5f}, rel_err_b={:.5f}'.format(rel_err_a, rel_err_b))
```

Messung A: $ 80.02cal/g \pm 8 \cdot 10^{-3} \% $

Messung B: $ 79.98cal/g \pm 14 \cdot 10^{-3} \% $

# 4.6

$$ X_n \sim \mathcal{U}([0, 1]) $$

$$ \frac{b-a}{n} \sum_{i=1}^{n} f(X_i) $$

gesucht: Fläche von Integral (Glockenkurve: Normalverteilung!) $$ I(f) = \frac{1}{\sqrt{2\pi}} \int_0^1 e^{\frac{-x^2}{2}} \mathcal{d} x $$

```python
import numpy as np
from scipy.stats import uniform, norm
from math import sqrt, exp, pi

def f(x):
    return 1 / sqrt(2 * pi) * np.exp(np.negative(x**2)/2)

xs = uniform.rvs(loc=0, scale=1, size=1000)
ys = f(xs)

area = (1-0)/len(xs) * ys.sum()
print('estimated: {:.3f}'.format(area))

area = norm.cdf(1) - norm.cdf(0)
print('calculated: {:.3f}'.format(area))
```
