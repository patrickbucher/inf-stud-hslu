## Aufgabe 4: Kantendetektion

```python
pic = 'DIP+MatLab-Gonzalesetal-F04.09a.tif'
image = cv2.imread(pic)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Original
plt.subplot(1,5,1)
plt.imshow(image, cmap='gray')

# Canny: Ergibt eine rein binÃ¤re Maske (schwarz/weiss), keine Abstufungen.
canny = cv2.Canny(image, 100, 200)
plt.subplot(1,5,2)
plt.imshow(canny, cmap='gray')

# Laplace: Mit Linien trukturierte FlÃ¤chen sind gut erkennbar.
laplace = cv2.Laplacian(image, cv2.CV_64F)
plt.subplot(1,5,3)
plt.imshow(laplace, cmap='gray')

# Sobel X: Die vertikel verlaufenden Linien sind besser erkennbar.
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
plt.subplot(1,5,4)
plt.imshow(sobel_x, cmap='gray')

# Sobel Y: Die horizontal verlaufenden Linien sind besser erkennbar (v.a. Dach).
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
plt.subplot(1,5,5)
plt.imshow(sobel_y, cmap='gray')

plt.show()
```

## Aufgabe 5: Laplacefilter

```python
pic = 'moon.tif'
image = cv2.imread(pic)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')

w1 = np.matrix([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
img1 = cv2.filter2D(image, -1, w1)
plt.subplot(1,3,2)
plt.imshow(cv2.add(image, img1), cmap='gray')

w2 = np.matrix([[1, 1, 1], [1, -6, 1], [1, 1, 1]])
img2 = cv2.filter2D(image, -1, w2)
plt.subplot(1,3,3)
plt.imshow(cv2.add(image, img2), cmap='gray')

plt.show()
```

## Aufgabe 6: Mask Median Filter

```python
pic = 'DIP+MatLab-Gonzalesetal-F04.09a_scratched.tif'
image = cv2.imread(pic)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')

median = cv2.medianBlur(image, 5)
plt.subplot(1,2,2)
plt.imshow(median, cmap='gray')

plt.show()
```
